### data/yesterday.txt. 번역
1. txt 파일 읽기
2. request로 http 통신
3. 번역결과를 파일로 쓰기


```python
from requests import Request, Session

# 파일읽기
def get_text_list():
    result_list = []
    with open('data/yesterday.txt', 'r', encoding='utf8') as file:
        contents = file.read()
        result_list = contents.split('\n')
        #split() : 어떤 기준으로 문자열 나눌때
        #split(sep="", maxsplit="1") 공백기준으로 나누고, 문자열 1번쪼갬
    return result_list

# 번역된 결과 파일로 쓰기
def save_to_file(mylist):
    with open('data/yesterday_trans.txt', 'w', encoding='utf8')as file:
        file.writelines(mylist)
        # 1. 문자열로 이루어진 리스트를 텍스트 파일에 쓰기
        # - 한줄씩 쓰기 : write() -한꺼번에 모든 줄쓰기: writelines()
        # 2. 텍스트 파일 읽어오기
        # - 텍스트 한줄씩 읽어오기 : readline(), -모두 읽어: readlines()
# http통신 
def main():
    
    session =Session()
    
    client_id="_n9QuV_IqzYSI83XNtBh"
    client_secret="G3U0tiOpjO"

    url = "https://openapi.naver.com/v1/papago/n2mt"
    #enctext = "yesterday all my troubles seemed so far away"

    #dict
    req_headers = {
        "X-Naver-Client-Id":client_id,
        "X-Naver-Client-Secret":client_secret
    }
    
    ####여기부터 하기
    lyric_list = get_text_list()
    #list comprehension
    lyric_list = [lyric for lyric in lyric_list if len(lyric)!=0]# 공백만 체크해서 안나오게 
    print(lyric_list, len(lyric_list))
    # get_text_list()로 한줄씩 받아온거 lyric_list로 공백만 체크해 묶어
    print('---------------------------------------')
    
    trans_list= [] 
    for en_text in lyric_list:
        params = {
            'source':'en',
            'target':'ko',
            'text':en_text #변수니까 얘는
        }
        print(en_text)#이거 안하면 한국어 번역 안됨
        req = Request('POST', url, data=params, headers=req_headers)
        prepared = req.prepare()
        res = session.send(prepared)
        print(res.status_code)

        try:
            ko_text=res.json()['message']['result']['translatedText']
        except Exception as err:
            print(err)
        else:
            print(ko_text)
            
        trans_list.append(en_text+'\n')
        trans_list.append(ko_text+'\n')
    save_to_file(trans_list)    
    print('번역 종료')

main()
```

# req 객체
* req.body : POST 정보
* req.query : GET 정보, url로 전송된 쿼리 스트링 파라미터 담고 있음
* req.params : 내가 이름 붙인 라우트 파라미터 정보
* req.headers : http의 header 정보 가짐

# res 객체
* res.send : 다양한 유형의 응답 전송
* res.redirect : 브라우저 다이렉트
* res.json : json 응답 전송
