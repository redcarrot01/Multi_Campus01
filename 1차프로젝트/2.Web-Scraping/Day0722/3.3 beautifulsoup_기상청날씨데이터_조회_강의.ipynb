```python
import requests
from bs4 import BeautifulSoup
url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
html = requests.get(url).text
# BeautifulSoup으로 분석하기
soup = BeautifulSoup(html, 'html.parser')
# 원하는 데이터 추출하기
#soup.find("title").text 도 사용가능
title = soup.find("title").string
wf = soup.find("wf").string
print(title)
print(wf)

print(soup.find('location').find('province').text)
#print(soup.find('location',attrs={"wl_ver" : "3"}).find('province').text)
print(soup.find('location',attrs={"wl_ver" : "3"}).find('city').text)
datas =soup.find('location',attrs={"wl_ver" : "3"}).find_all("data")
# print(datas)
#총 데이터 13개인데 ..
data_list=[]
for data in datas:
    data_dict = {}
    data_dict['wf']=data.find('wf').text #{'wf': 'aaa'}
    data_dict['tmef']=data.find('tmef').text
    data_dict['reliability']=data.find('reliability').text
    print(data_dict)
    data_list.append(data_dict)
    #print(data)
   # print(data.find('wf').text,data.find('tmef').text,data.find('reliability').text)
data_list
```

```python
#전국의 날씨를 dict와 list 에 저장한다.
loc_list = []
for location in soup.find_all('location',attrs={"wl_ver" : "3"}):
    loc_dict = {}
    #print(len(location), location)
    loc_dict['province'] = location.find('province').text
    loc_dict['city'] = location.find('city').text
    data_list = []
    for data in location.find_all("data"):
        data_dict = {}
        data_dict['mode'] = data.find('mode').text
        data_dict['wf'] = data.find('wf').text
        data_dict['tmn'] = data.find('tmn').text
        data_dict['tmx'] = data.find('tmx').text
        data_list.append(data_dict)
    loc_dict['data'] = data_list    
    #print(loc_dict)
    #print('-----')
    loc_list.append(loc_dict)
loc_list
```

```python
# loc_list를 json 파일로 저장하기
import json

with open("data/weather.json", "w") as f:
    json.dump(loc_list, f)
```
