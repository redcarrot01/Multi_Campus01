### python.org -> pypi -> 다시 홈 되는거 해보기
### 웹드라이버 객체 생성

```python
from selenium import webdriver
import time

#webdriver 객체 생성(chromdriver.exe 를 생성)
browser = webdriver.Chrome('./chromedriver.exe')
#browser=> selenium.webdriver.chrome.webdriver.WebDriver (session="da67f30aa56a72c518a9bd79e76d96f0")>

#python.org요청
browser.get('https://www.python.org/') #브라우저가 새창에 뜸

#pypi 에 해당 element를 찾기 , #li여러개니까 elements
menus = browser.find_elements_by_css_selector('#top ul.menu li')

pypi = None
for menu in menus:
    print(menu.text)
    if menu.text == 'PyPI':
        pypi = menu
#click() 함수 - 클릭?? 왜?
pypi.click()
#sleep
time.sleep(3)
#driver 종료
browser.quit()
```
